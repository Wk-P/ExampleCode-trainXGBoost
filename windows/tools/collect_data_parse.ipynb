```python
# read from local file

from pathlib import Path
import json
import typing

def read_data(file_path: typing.Union[str, Path]) -> typing.Any:
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    return data

def parse_data(data):
    parsed = []
    for item in data:
        parsed.append({
            'input': item.get('input', ''),
            'expected_output': item.get('expected_output', '')
        })
    return parsed

def view_data(data: typing.Any):
    print(data)
```

```python

data_folder = Path.cwd() / 'data'

# get all names of files in data folder
data_files = [f for f in data_folder.iterdir() if f.is_file()]

print(f"Found {len(data_files)} data files in {data_folder}.")

data_list = []
# read from folder data
for data_file in data_files:
    data = read_data(data_file)
    data_list.append({
        'file_name': data_file.name,
        'data': data
    })
print(f"{'request_numbers':<20}   processing_times\n")
for index in range(len(data['request_numbers'])):
    print(f"{data['request_numbers'][index]:<20} | {data['response_processing_times'][index]}")
```

```python
# read all data file and count all data lines
total_lines = 0
all_records = dict()
for data_item in data_list:
    total_lines += len(data_item['data']['request_numbers'])
    if all_records.get('request_numbers') is None:
        all_records['request_numbers'] = data_item['data']['request_numbers']
        all_records['response_processing_times'] = data_item['data']['response_processing_times']
    else:
        all_records['request_numbers'].extend(data_item['data']['request_numbers'])
        all_records['response_processing_times'].extend(data_item['data']['response_processing_times'])
print(f"Total data lines across all files: {total_lines}")

# save all data to a new file
all_data_file = data_folder / 'all_data.json'
with open(all_data_file, 'w', encoding='utf-8') as f:
    json.dump(all_records, f, ensure_ascii=False, indent=4)

print(f"All data saved to {all_data_file}")
```
