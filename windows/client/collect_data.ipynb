```python
# read data from '*CPU*' json files
# read data from '*HDD*' json files
# read data from '*MEM*' json files
import json

all_files: list[Path]

cpu_data = dict()
hdd_data = dict()
mem_data = dict()
for file in all_files:
    if 'CPU' in file.name:
        with open(file, 'r') as f:
            data = dict(json.load(f))
            cpu_data = cpu_data | data
    elif 'HDD' in file.name:
        with open(file, 'r') as f:
            data = dict(json.load(f))
            hdd_data = hdd_data | data
    elif 'MEM' in file.name:
        with open(file, 'r') as f:
            data = dict(json.load(f))
            # combine 'MEM' data from all '*MEM*' files
            for key, value in data.items():
                if key in mem_data:
                    mem_data[key].extend(value)
                else:
                    mem_data[key] = value

print([k for k in cpu_data.keys()])
print([k for k in hdd_data.keys()])
print([k for k in mem_data.keys()])

print(len(cpu_data.get('processing_time', [])))
print(len(hdd_data.get('processing_time', [])))
print(len(mem_data.get('processing_time', [])))
```

```python
# print sample data
sample_size = 2
print("CPU Data Sample:")
for key, value in cpu_data.items():
    print(f"{key}: {value[:sample_size]}")

print("\nHDD Data Sample:")
for key, value in hdd_data.items():
    print(f"{key}: {value[:sample_size]}")

print("\nMEM Data Sample:")
for key, value in mem_data.items():
    print(f"{key}: {value[:sample_size]}")
```
